# Get play by play for game

```python
from nba_api.stats.endpoints import GameRotation
from nba_api.stats.endpoints import BoxScoreTraditionalV2
```

```python
box_score = BoxScoreTraditionalV2(game_id= game_ids[0]).get_data_frames()[0]
```

```python
box_score.columns
```

```python
box_score['LAST_NAME'] = box_score.apply(lambda row: row['PLAYER_NAME'].replace(row['NICKNAME'], '').strip(), axis = 1)
player_dictionary = {}
for index, row in box_score.iterrows():
    team_id = row['TEAM_ID']
    player_id = row['PLAYER_ID']
    last_name = row['LAST_NAME']
    if team_id in player_dictionary:
        if player_id in player_dictionary[team_id]:
            continue
        else:
            player_dictionary[team_id][player_id] = last_name
    else:
        player_dictionary[team_id] = {}
        player_dictionary[team_id][player_id] = last_name
```

```python
player_dictionary
```

```python
game_rotation = GameRotation(game_id= game_ids[0]).get_data_frames()
game_rotation_home, game_rotation_away = game_rotation[0], game_rotation[1]
game_rotation_home.to_csv('game_rot.csv')
```

```python
home_lineup = list(game_rotation_home[game_rotation_home['IN_TIME_REAL'] == 0]['PERSON_ID'])
away_lineup = list(game_rotation_away[game_rotation_away['IN_TIME_REAL'] == 0]['PERSON_ID'])
```

```python
columns = ["lineup_ID"] + [f"player_{team}_{n}" for team in ['home', 'away'] for n in range(1, 6)] + ["net_rating", "time"]
```

```python
df_net_rating = pd.DataFrame(columns=columns)
```

```python
lineup = home_lineup + away_lineup
lineup_ID = str(sorted(lineup))
lineup_data = pd.Series([lineup_ID] + lineup, index = columns[:-2])
```

```python
from nba_api.stats.endpoints import PlayByPlayV3
```

```python
plays = PlayByPlayV3(game_id= game_ids[0]).get_data_frames()[0]
```

```python
plays.columns
```

```python
plays['actionType'].unique()
```

```python
plays[['scoreHome', 'scoreAway']] = plays[['scoreHome', 'scoreAway']].replace('', np.nan)
plays[['scoreHome', 'scoreAway']] = plays[['scoreHome', 'scoreAway']].ffill()
```

```python
substitutions = plays[plays['actionType'] == 'Substitution'].copy()
```

```python
substitutions['scoreDiff'] = substitutions['scoreHome'].astype(int) - substitutions['scoreAway'].astype(int)
substitutions['netDiff'] = substitutions['scoreDiff'].diff().fillna(substitutions['scoreDiff'])
```

```python
df_net_rating['net_rating'] = substitutions['netDiff']
df_net_rating.iloc[0, :11] = lineup_data
```

```python
substitutions.head()
```

```python
substitutions['sub_coming_in'] = substitutions['description'].str.replace('SUB: ', '').str.strip().str.split(' ').str[0]
```
