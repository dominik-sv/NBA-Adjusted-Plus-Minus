# Get game ids

```python
from nba_api.stats.endpoints import LeagueGameLog
import pandas as pd
import numpy as np
```

```python
game_log = LeagueGameLog().get_data_frames()[0]
```

```python
df = pd.DataFrame(game_log)
```

```python
game_ids = df['GAME_ID'].unique()
```

# Get play by play for game

```python
from nba_api.stats.endpoints import PlayByPlayV3
```

```python
help(PlayByPlayV3)
```

```python
plays = PlayByPlayV3(game_id= game_ids[0]).get_data_frames()[0]
```

```python
plays.columns
```

```python
plays['actionType'].unique()
```

```python
substitutions = plays[plays['actionType'] == 'Substitution']
```

```python
players = plays['playerNameI'].unique()[1:]
```

```python
print(players)
```

```python
for player in players:
    player_subs = substitutions[substitutions['playerNameI'] == player]
    print(player_subs['scoreHome'])
    print(player_subs['scoreAway'])
    break
```
